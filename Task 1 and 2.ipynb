```python
#Mapping each attribute to an attack type. We have 4 attack types and a normal category
mapped_attack = [x.strip().split() for x in open('Attack_types.txt', 'r')]
mapped_attack = {a:b for (a,b) in mapped_attack}
```

```python
#Visualizing our attack map
mapped_attack
```

```python
#Replacing the attack_category from the datafram of Dataset by the types of attacks in the attack_map
df['attack_category'] = df['attack_category'].replace(mapped_attack)
```

```python
#encoding the categorical values from the dataframe and replacing them with dummy numerical values
def cat_encoding(df1, col1):
    return pd.concat([df1.drop(col1, axis=1), pd.get_dummies(df1[col1].values)], axis=1)
```

```python
#Some of the values in the dataframe are very wide in range. Such as duration, Source bytes and Desination Bytes.
#To concatenate them in a small range, we do the log scalling.
def log_scalling(df1, col1):
    return df1[col1].apply(np.log1p)
```

```python
#Catagorical Values
cat_list = ['protocol_type', 'service', 'flag']
for col in cat_list:
    df = cat_encoding(df, col)
```

```python
#Log scalling values
log_list = ['duration', 'src_bytes', 'dst_bytes']
for col in log_list:
    df[col] = log_scalling(df, col)
```

```python
#Saving the attack_category labels in target variable and dropping the values in the column afterwards.
target_var = df.pop('attack_category')
```

```python
#Getting dummy values
target_var = pd.get_dummies(target_var)
```

```python
#Preparing two variables for training of our model.
target_var = target_var.values
train = df.values
```

```python
#Transform all the features in the range of 0 and 1 which is the default value of MinMaxScaler
min_max_scaler = MinMaxScaler()
train = min_max_scaler.fit_transform(train)
```

```python
train.shape
```

```python
#Importing all the required libraries for building the model.
from keras.callbacks import EarlyStopping
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.layers.embeddings import Embedding
```

**This is our Model 1**
It has 6 total layers. All the layers have ReLU activation except for the last one which has a 5 way sofmax activation function because we have 5 classes to predict.
The 1st layer has 128 neurons.
The 2nd layer has 16 neurons.
The 3rd layer has 256 neurons.
The 4th layer has 128 neurons.
The 5th layer has 32 neurons.
The 6th layer has 5 neurons.
**The Batch size is 128 and the validation split used is 0.1.**
**The optimizer used is Adam.**
**Epochs are 10**

```python
#Building the network through keras framework
def build_network():
    models = []
    model = Sequential()
    model.add(Dense(128, input_dim=123))
    model.add(Activation('relu'))
    model.add(Dense(16))
    model.add(Activation('relu'))
    model.add(Dense(256))
    model.add(Activation('relu'))
    model.add(Dense(128))
    model.add(Activation('relu'))
    model.add(Dense(32))
    model.add(Activation('relu'))
    model.add(Dense(5))
    model.add(Activation('softmax'))

    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

    return model
```

```python
#Early stopping
network = build_network()
early_stopping = EarlyStopping(monitor='val_loss', min_delta=0, patience=3, verbose=0, mode='auto')
```

```python
#Runnig the model
network.fit(x=train, y=target_var, epochs=10, validation_split=0.1, batch_size=128, callbacks=[early_stopping])
```
